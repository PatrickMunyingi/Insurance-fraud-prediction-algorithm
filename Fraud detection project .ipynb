```python
def clean_policy_state(x):
    if 'OH' in x:
        return 'Ohio'
    if 'IL' in x:
        return 'Illionis'
    if 'IN' in x:
        return 'Indiana'
    
claim['policy_state']=claim['policy_state'].apply(clean_policy_state)
```

```python
def clean_incident_state(x):
    if 'NY' in x:
        return 'NewYork'
    if 'SC' in x:
        return 'South Carolina'
    if 'WV' in x:
        return 'West Virginia'
    if 'NC' in x:
        return 'North Carolina'
    if 'PA' in x:
        return 'Pennsylvania'
    if 'OH' in x:
        return 'Ohio'
claim['incident_state']=claim['incident_state'].apply(clean_incident_state)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
claim['fraud_reported']=le.fit_transform(claim['fraud_reported'])
```

```python
data=pd.DataFrame(data,columns=['years_as_customer', 'age', 'policy_state', 'policy_csl',
       'policy_deductable', 'policy_annual_premium', 'insured_sex',
       'insured_education_level', 'insured_occupation', 'insured_hobbies',
       'insured_relationship', 'incident_type', 'collision_type',
       'incident_severity', 'incident_state', 'number_of_vehicles_involved',
       'property_damage', 'bodily_injuries', 'injury_claim', 'property_claim',
       'vehicle_claim', 'auto_make', 'auto_year'])
data.head()
```

```python
claims=pd.concat([data,claim['fraud_reported']],axis=1)
claims.head()
```

# Model Building

```python
X=claims.drop('fraud_reported',axis=1)
y=claims['fraud_reported']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1,stratify=y, random_state=42)
```

```python
lr.fit(X_train,y_train)
```

```python
pred=lr.predict(X_test)
print(pred)
```

# Model Evaluation

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score
```

```python
print("Our Evaluation Metrics are:")
print('The Confusion matrics')
print(confusion_matrix(y_test,pred))
print("The accuracy,precision and recall score:")
print(accuracy_score(y_test,pred))
print(precision_score(y_test,pred))
print(recall_score(y_test,pred))
```
